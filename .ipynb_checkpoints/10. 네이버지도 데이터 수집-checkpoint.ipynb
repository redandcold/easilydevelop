```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
#keys빼고는 필수
```

```python
driver=wb.Chrome()
driver.get("https://map.naver.com/p?c=15.00,0,0,0,dh")
#광고 꺼야함
```

```python
# 검색창에 검색어를 입력
# 네이버지도는 검색창 id가 켤 때마다 바뀐다.
search=driver.find_element(By.CSS_SELECTOR,".input_search")
search.send_keys("동명동 맛집")
search.send_keys(Keys.ENTER)
```

```python
#가게 이름 수집
#접근이 불가능한 이유는 iframe으로 불러와서
title_list=driver.find_elements(By.CSS_SELECTOR, "span.place_bluelink.TYaxT")
title_list[0].click()
```

```python
#1. 해당하는 iframe으로 이동, 시야를 이동
driver.switch_to.frame("searchIframe")
```

```python
#세부사항에서 가게 이름정보 수집
name=driver.find_element(By.CSS_SELECTOR, "#_title > span.Fc1rA")
name.text
```

```python
# 다른 iframe으로 이동
# 한번 다른 iframe으로 이동했다면 원본 iframe으로 돌아간 후 원하는 frame으로 이동
driver.switch_to.default_content()
#실행하면 화면엔 변화가 없지만 컴퓨터 시야는 바뀜
```

```python
# 스크롤을 이동
# 화면이 분리되어 위에는 고정되어 있고 밑에는 스크롤이 된다.
# 스크롤 내릴 영역을 먼저 선택
# 시야는 한번만 바꾼다.
div=driver.find_element(By.CSS_SELECTOR,"div.XUrfU")
div.click()
for i in range(10):
    
    body=driver.find_element(By.CSS_SELECTOR, "body")
    body.send_keys(Keys.END)
    
```

# Selenium에서 iframe 조작하는 법
- selenium을 통해서 iframe을 접근하려면 반드시 바라보는 frame을 이동해야 한다.
- driver.switch_to.frame("frameID")
- 원본 데이터로 이동= driver.switch_to.default_content()
- ★프레임끼리는 이동이 불가-> 프레임 이동한 후->원본으로 돌아오고-> 다른 프레임으로 이동
- 프레임 속에 스크롤을 조작하는 경우
1. 모든 데이터가 body 안에서 동작한다면 (화면에 분할이 없는 경우)->스크롤 진행가능
2. body 안에 화면이 분할된 경우 -> 분할된 부분을 클릭한 다음-> 스크롤 진행
