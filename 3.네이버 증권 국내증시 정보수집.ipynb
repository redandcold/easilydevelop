```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
#1. 네이버 시가총액 사이트 정보 요청
res=req.get("https://finance.naver.com/sise/sise_market_sum.naver")
res
```

```python
#2. 응답코드를 확인하고, 정상적인 응답이면 bs통해서 데이터 변형
soup=bs(res.text, "lxml")
soup
```

```python
#3. 변형된 데이터에서 종목명, 현재가격 정보를 수집
top50=soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr > td > a")
print(top50)
soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr > td > a")[0].text
```

```python
soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr > td:nth-child(2) > a")# 행 전체 정보를 받아오려면 tr 뒤 nth지운다
```

```python
price_list=[]
price=soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr > td:nth-child(3)")
for i in price:
    price_list.append(i.text)
price_list
```

```python
#4. 순수한 텍스트 정보로 변형작업(리스트를 제작)
top50_list=[]

for i in top50:
    top50_list.append(i.text)
top50_list
```

```python
top50_list1=[]
for i in top50_list:
    if i=='':
        continue
    else:
        top50_list1.append(i)
len(top50_list1)
```

```python
#5. 표를 제작, 한글이 깨지지 않게 csv파일로 저장
import pandas as pd
```

```python
rank_list=[]
for i in range(1,51):
    rank_list.append(i)
dic={"순위":rank_list, "기업명":top50_list1}
df_top50=pd.DataFrame(dic)
```

```python
df_rank50=df_top50.set_index("순위")
```

```python
df_rank50["현재가"]=price_list
df_rank50
```

```python
df_rank50.to_csv("시가총액순위.csv", encoding="utf-8")
```
