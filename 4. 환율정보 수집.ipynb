# iframe 데이터처리
- 화면 상에 정보는 보이지만 실제 데이터는 다른 공간에 존재할 때 사용

```python
# iframe은 다른 공간에 존재하는 정보를 화면에 표출만 해준다.
# 해결법 = 실제 데이터가 존재하는 URL로 요청을 보내야한다.
# iframe 태그의 src 안에 있는 경로를 사용하면 된다.
# 대부분의 경로는 상대경로로 작성이 되어있다.
# 상대경로를 절대경로와 합쳐서 URL 제작
res=req.get("https://finance.naver.com/marketindex/exchangeList.naver")
res
soup=bs(res.text, "lxml")
soup
money=soup.select("body > div > table > tbody > tr > td.tit > a")
price=soup.select("body > div > table > tbody > tr > td.sale")
print(len(price), len(money))
```

```python
#개행문자 혹은 공백 등을 지울 때는 strip()활용
money_list=[]
price_list=[]
rank_list=[]
for i in range(len(price)):
    money_list.append(money[i].text.strip())
    price_list.append(price[i].text)
    rank_list.append(i+1)
print(rank_list,money_list, price_list)
```

```python
import pandas as pd
```

```python
dic={"순번":rank_list,"통화명":money_list,"기준가(원)":price_list}
df_currency=pd.DataFrame(dic)
df_currency
```

```python
table_currency=df_currency.set_index("순번")
table_currency
```

```python
table_currency.to_csv("종합환율정보.csv", encoding="utf-8")
```
