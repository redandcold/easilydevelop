```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
import math
import re
# 리뷰 총 개수 입력받아서 끝나면 전 페이지로 넘어가기

driver=wb.Chrome()
driver.get("https://prod.danawa.com/list/?cate=18242367&15main_18_02")
time.sleep(1)
#상품명
#product=driver.find_elements(By.CSS_SELECTOR, "div.prod_info > p.prod_name > a")

    #상품의견많은순 클릭
driver.find_element(By.CSS_SELECTOR, "#productListArea > div.prod_list_opts > div.order_opt > ul > li:nth-child(6) > a").click()
time.sleep(2)
while(True):
   # try: #상품목록 한 페이지씩 넘기다가 마지막페이지 이후에 뜨는 오류 잡기
    for l in range(10): # 상품목록에서 페이지 하나씩 넘길 반복문
        product=driver.find_elements(By.CSS_SELECTOR, "div.prod_info > p.prod_name > a"
        for k in range(len(product)): #상품 하나씩 선택할 반복문
            product_name=product[k].text
            print(product_name)
            product[k].click() #새탭열림 상품하나 선택
            time.sleep(1)


            title_list=[]
            content_list=[]

            #새탭에 접근
            driver.switch_to.window(driver.window_handles[-1])
            time.sleep(1)

            #리뷰개수로 페이지10개짜리 넘기는 빈도 알아내기
            page=0
            
            
            reviewcount=driver.find_element(By.CSS_SELECTOR,"#productOpinionTabCount").text
            review_count=int(re.sub(r'[^0-9]','',reviewcount))
            if review_count%10==0:
                page=review_count/10
            else:
                page=review_count/10+1
            next10_count=int((page-1)/10)


            review_button= driver.find_element(By.CSS_SELECTOR, "#bookmark_cm_opinion_item > a")
            #후기 버튼 누르기
            review_button.click()
            time.sleep(0.5)
            print(next10_count)
            num=1 #현재 페이지 수
            for tenpage in range(next10_count): #10페이지씩 넘기는 수만큼 반복
                #try: # 끝 페이지 도달 시 range 오류가 뜰 것이고 이때 종류
                if num==1:
                    print(num, end="\t")
                    num+=1
                    tit=driver.find_elements(By.CSS_SELECTOR,"p.tit")
                    for j in range(len(tit)):
                        tit=driver.find_elements(By.CSS_SELECTOR,"p.tit")
                        con=driver.find_elements(By.CSS_SELECTOR,"div.atc")
                        title_list.append(tit[j].text)
                        content_list.append(con[j].text)

                else:
                    for i in range(1,10):
                        #하나씩 10페이지 넘기기
                        tit=driver.find_elements(By.CSS_SELECTOR,"p.tit")
                        con=driver.find_elements(By.CSS_SELECTOR,"div.atc")

                        num+=1
                        print(num, end="\t")
                        for j in range(len(tit)):
                            title_list.append(tit[j].text)
                            content_list.append(con[j].text)
                        try:
                            driver.find_elements(By.CSS_SELECTOR,".page_num")[i].click()
                            time.sleep(0.5)
                        except:
                            break


                #10페이지 단위 넘기기
                driver.find_elements(By.CSS_SELECTOR,"div.page_nav_area >a")[-1].click()
                time.sleep(1)
                print(num, end="\t")
               # except: #csv파일 만들어주고 상품페이지에서 돌린 while문 탈출
            num_list=[]
            for i in range(len(title_list)):
                num_list.append(i+1)
            dic={"순번":num_list,"제목":title_list, "후기":content_list}
            df_review=pd.DataFrame(dic)
            dfreview=df_review.set_index("순번")
            dfreview.to_csv(f"{product_name}_다나와리뷰.csv", encoding="utf-8")
            driver.switch_to.window(driver.window_handles[0])

        nextlist=driver.find_elements(By.CSS_SELECTOR,"a.num")
        nextlist[l].click()#상품목록 다음페이지 하나씩 클릭

    next_ten=driver.find_elements(By.CSS_SELECTOR, "#productListArea > div.prod_num_nav > div > a") 
    next_ten.click()#상품목록 10페이지씩 넘기기
#     except:
#         break
        
```
